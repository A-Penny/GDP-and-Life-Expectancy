# GDP and Life Expectancy
### This project attempts to convey the relationship, if any, between a country's GDP and life expectancy. 

## The first thing to look at is a sample of the data that was acquired.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

all_data = pd.read_csv('all_data.csv')
print(all_data.head())
print(len(all_data))
```

The data for this project is acquired from the following sources:\
World Bank ( for GDP )\
World Health Organization ( for life expectancy )


```python
countries = all_data['Country'].unique()
years = all_data['Year'].unique()
print('There are six countries in the sample study:\n')
for i in range(len(countries)):
    print(str((i + 1)) + ': ' + str(countries[i]) + ' (years ' + str(years.min()) + '-' + str(years.max()) + ')')
    
```

## Here is the change in life expectancy and gdp for each country from 2000-2015:

```python
countries_life = []
chile_life = all_data[all_data.Country == 'Chile']['Life expectancy at birth (years)']
china_life = all_data[all_data.Country == 'China']['Life expectancy at birth (years)']
germany_life = all_data[all_data.Country == 'Germany']['Life expectancy at birth (years)']
mexico_life = all_data[all_data.Country == 'Mexico']['Life expectancy at birth (years)']
usa_life = all_data[all_data.Country == 'United States of America']['Life expectancy at birth (years)']
zimbabwe_life = all_data[all_data.Country == 'Zimbabwe']['Life expectancy at birth (years)']
countries_life.append(chile_life)
countries_life.append(china_life)
countries_life.append(germany_life)
countries_life.append(mexico_life)
countries_life.append(usa_life)
countries_life.append(zimbabwe_life)


for country in countries_life:
    plt.plot(range(len(country)), country, marker = 'o')
plt.legend(countries)
ax = plt.subplot()
ax.set_xticks(range(len(china_life)))
ax.set_xticklabels(years)
plt.xticks(rotation = 45)
plt.xlabel('Year')
plt.ylabel('Age (in years)')
plt.title('Life Expectancy from 2000 - 2015')
plt.show()
plt.clf()
```

```python
countries_gdp = []
chile_gdp = all_data[all_data.Country == 'Chile'].GDP
china_gdp = all_data[all_data.Country == 'China'].GDP
germany_gdp = all_data[all_data.Country == 'Germany'].GDP
mexico_gdp = all_data[all_data.Country == 'Mexico'].GDP
usa_gdp = all_data[all_data.Country == 'United States of America'].GDP
zimbabwe_gdp = all_data[all_data.Country == 'Zimbabwe'].GDP
countries_gdp.append(chile_gdp)
countries_gdp.append(china_gdp)
countries_gdp.append(germany_gdp)
countries_gdp.append(mexico_gdp)
countries_gdp.append(usa_gdp)
countries_gdp.append(zimbabwe_gdp)


for country in countries_gdp:
    plt.plot(range(len(country)), country / 1000000000000, marker = 'o')
    
plt.legend(countries)
ax = plt.subplot()
ax.set_xticks(range(len(china_gdp)))
ax.set_xticklabels(years)
plt.xticks(rotation = 45)
plt.xlabel('Year')
plt.ylabel('GDP (Trillion USD)')
plt.title('GDP from 2000 - 2015')
plt.grid(alpha = 0.2)
plt.show()
plt.clf()
```

As you can see in the graph above, the United States had a vastly larger GDP than the other countries in the sample. China also showed remarkable growth over the years in the study. The other countries in the study did not exhibit gdp growth from 2000 - 2015.

## Let's take a look at each country's average life expectancy versus its average GDP.

```python
countries_life_mean = []
countries_gdp_mean = []
for country in countries_life:
    x = np.mean(country)
    countries_life_mean.append(x)
for country in countries_gdp:
    x = np.mean(country) / 1000000000000
    countries_gdp_mean.append(x)
plt.figure(figsize = [10, 4])
plt.subplot(1, 2, 1)
plt.bar(countries, countries_life_mean)
plt.xticks(rotation = 75)
plt.ylim(40, 90)
plt.title('Average Life Expectancy by Country from 2000 - 2015')
plt.xlabel('Country')
plt.ylabel('Age in Years')
plt.subplot(1, 2, 2)
plt.bar(countries, countries_gdp_mean)
plt.xticks(rotation = 75)
plt.xlabel('Country')
plt.ylabel('Average GDP (trillion USD)')
plt.title('Average GDP by Country from 2000 - 2015')
plt.show()
```

### Based on the results of the study, it appears that a country's GDP does not have a correlation to its average life expectancy. Zimbabwe has a negligible GDP and a very low comparitive life expectancy. However, Chile and Germany are among the lowest in terms of GDP, but they have the highest life expectancy. As a conclusion, the findings of this study would warrant a different exploration as to a possible cause for low life expectancy.
